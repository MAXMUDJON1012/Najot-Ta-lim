```python
def eng_katta_son(*sonlar):
    if not sonlar:
        return None
    return max(sonlar)
natija1 = eng_katta_son(12, 45, 2, 89, 34)
print(f"Eng katta son: {natija1}")  

natija2 = eng_katta_son(-5, -23, -1, -12)
print(f"Eng katta son: {natija2}")  
```
