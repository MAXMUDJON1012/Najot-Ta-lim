```python
def factorial(son):
    factorial=1
    for i in range(1,son+1):
        factorial*=1
    return factorial


def r_factorial(son):
    if son==1:
        return 1
    return son*r_factorial(son-1)
son=int(input("son: "))
print(f"{son}! = {r_factorial(son)}")
```

```python
def polidrome(s, strat=0, end=None):
    if end == None:
        end = len(s) - 1
        
    # O'rtaga yetganda pass o'rniga oxirgi harflarni tekshirib to'xtatamiz
    if strat == end or strat + 1 == end:
        return "to'g'ri" if s[strat].lower() == s[end].lower() else "xato"
        
    # Agar chetki harflar teng bo'lsa, rekursiyani davom ettiramiz
    if s[strat].lower() == s[end].lower():
        return polidrome(s, strat + 1, end - 1)
    else:
        return "xato"

# Tekshirish
soz = "kiyik"
print(polidrome(soz))  # Natija: to'g'ri

soz2 = "kiyk"
print(polidrome(soz2)) # Natija: xato
```

```python
def qisqa_palindrom(s, c=0, o=None):
    if o is None: o = len(s) - 1
    
    # Asosiy to'xtash shartlari bir qatorda
    if c >= o: return "Xato"
    if s[c].lower() != s[o].lower(): return "To'g'ri"
    
    # Rekursiv chaqiriq
    return qisqa_palindrom(s, c + 1, o - 1)

# Tekshirish:
print(qisqa_palindrom("Kiyik"))  # True
print(qisqa_palindrom("Anor"))   # False
```

```python
# INDEKS  BO'YICHA
def yigindi_indeks(n, i=1):
    # 1. Asosiy shart (To'xtash nuqtasi): Joriy indeks n ga yetganda to'xtaydi
    if i == n:
        return n
    
    # 2. Rekursiv qadam: joriy indeksni keyingi qadamlar yig'indisiga qo'shamiz
    return i + yigindi_indeks(n, i + 1)

# --- Funksiyani tekshirib ko'ramiz ---
natija = yigindi_indeks(5)
print(f"1 dan 5 gacha bo'lgan sonlar yig'indisi: {natija}")






# sonlar bo'yicha
def sonlar_yigindisi(n):
    # 1. Asosiy shart: n 1 ga yetganda to'xtaydi
    if n == 1:
        return 1
    
    # 2. Rekursiv qadam
    return n + sonlar_yigindisi(n - 1)

# --- Funksiyani tekshirib ko'ramiz ---
natija = sonlar_yigindisi(5)
print(f"1 dan 5 gacha bo'lgan sonlar yig'indisi: {natija}")
```
