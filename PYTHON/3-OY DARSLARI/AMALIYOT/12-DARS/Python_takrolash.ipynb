```python
# 3
price = []
for i in range(1, 6):
    narx = int(input("enter the price !"))
    price.append(narx)

eng_katta = price[0]
eng_kichik = price[0]
yigindi = 0
elementlar_soni = 0

for s in price:
    if s > eng_katta:
        eng_katta = s
    if s < eng_kichik:
        eng_kichik = s
    yigindi += s
    elementlar_soni += 1
    
o_rtachasi = yigindi / elementlar_soni

print("Eng katta:", eng_katta)
print("Eng kichik:", eng_kichik)
print("O'rtacha:", o_rtachasi)
```

```python
# 6
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

for n in range(1, 51):
    if is_prime(n):
        print(n)
```

```python
# 14
def power(n,d):
    if d==1:
        return n
    return n*power(n,d-1)
print(power(2,5))
```

```python
# 18
def ism_tekshir(ism):
    uzunlik = len(ism)
    if uzunlik < 3 or uzunlik > 20:
        raise ValueError("Ism uzunligi 3 dan kichik yoki 20 dan katta bo'lishi mumkin emas!")
    return f"Salom, {ism}"

try:
    user_ism = input("Ismingizni kiriting: ")
    print(ism_tekshir(user_ism))
except ValueError as xato:
    print("Xatolik yuz berdi:", xato)
```
