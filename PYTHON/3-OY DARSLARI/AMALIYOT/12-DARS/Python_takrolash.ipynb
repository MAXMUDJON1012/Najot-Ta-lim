```python
price = []

for i in range(1, 6):
    narx = int(input("enter the price !"))
    price.append(narx)

eng_katta = price[0]
eng_kichik = price[0]
yigindi = 0
elementlar_soni = 0

for s in price:
    if s > eng_katta:
        eng_katta = s
    if s < eng_kichik:
        eng_kichik = s
    yigindi += s
    elementlar_soni += 1
    
o_rtachasi = yigindi / elementlar_soni

print("Eng katta:", eng_katta)
print("Eng kichik:", eng_kichik)
print("O'rtacha:", o_rtachasi)
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

for n in range(1, 51):
    if is_prime(n):
        print(n)
```
