```python
                                                                    # Masalalar: map, filter, reduce Exception Modullar
```

```python
numbers=[1,2,3,4]
square=map(lambda x:x**2,numbers)
print(list(square))
```

```python
text=["python", "kod"]
lenn=map(lambda x:len(x),text)
print(list(lenn))
```

```python
numbers=[5, 12, 8, 20]
higher_number=filter(lambda x: x > 10,numbers)
print(list(higher_number))
```

```python
text=["it", "kod", "python"]
lenn=filter(lambda x:len(x)>4,text)
print(list(lenn))
```

```python
teperature=[0, 100, 37]
farangate=map(lambda C: C * 9 / 5 + 32,teperature)
print(list(farangate))
```

```python
numbers=[1, 2, 3, 4]
pair_number=filter(lambda x : x % 2 ==0 , numbers)
square=map(lambda x : x**2 , pair_number)
print(list (square)) 
```

```python
from functools import reduce
sonalr = [ 3,5,4,31,6]
yigindi =reduce ( lambda a , b : a+b , sonalr)
katta=reduce ( lambda a , b : a if a>b else b ,sonalr)
print(yigindi)
print(katta)
```

```python
from functools import reduce


savatcha = [
    {
        "nomi": "Noutbuk",
        "narxi": 8500000,  # so'mda
        "soni": 1
    },
    {
        "nomi": "Sichqoncha (Mouse)",
        "narxi": 150000,
        "soni": 2
    },
    {
        "nomi": "Klaviatura",
        "narxi": 350000,
        "soni": 1
    }
]
def summa(a,mahsulot):
    return a + mahsulot["narxi"] * mahsulot["soni"]
jami_summa = reduce(summa,savatcha,0)
jami_summa2=reduce(lambda a,mahsulot: a + mahsulot["narxi"] * mahsulot["soni"],savatcha, 0)
print(jami_summa)
print(jami_summa2)
```
