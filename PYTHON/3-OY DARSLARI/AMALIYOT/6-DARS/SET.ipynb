```python
main_test = {
 "Ali", "Vali", "Hasan",
 "Aziz", "Diyor", "Jasur"
}
bonus_test = {
 "Hasan", "Aziz",
 "Kamol", "Jasur", "Asad"
}


asosiy_bonus=main_test.intersection(bonus_test)
print(asosiy_bonus)

only_main=main_test.difference(bonus_test)
print(only_main)

only_main2=bonus_test.difference(main_test)
print(only_main2)

atlist_one=main_test.symmetric_difference(bonus_test)
print(atlist_one)

atlist_one = main_test.union(bonus_test)
print(atlist_one)
```

```python
dc = {
 "Hasan", "Aziz",
 "Jasur", "Asad", "Kamol"
}
marvel = {
 "Ali", "Vali", "Hasan",
 "Aziz", "Kamol"
}
natija1 = marvel.intersection(dc)
natija2 = marvel.difference(dc)
natija3 = dc.difference(marvel)
natija4 = marvel.union(dc)
print(natija1,natija2,natija3,natija4)
```

```python
entered = {
 "Ali", "Vali", "Hasan",
 "Aziz", "Diyor", "Jasur"
}
left = {
 "Hasan", "Aziz",
 "Diyor", "Kamol", "Asad"
}


shart_1 = entered & left
shart_2 = entered - left
shart_3 = left - entered
shart_4 = entered | left
print("1. Ertalab kirib, kechqurun ham chiqqanlar:", shart_1)
print("2. Ertalab kirib, chiqishi qayd etilmaganlar:", shart_2)
print("3. Chiqqan, lekin kirishi qayd etilmaganlar:", shart_3)
print("4. Kun davomida qayd etilgan barcha o'quvchilar:", shart_4)
```
