```python
def juft_sonlar_yigindisi(n):
    if n < 2:
        return 0
    if n % 2 == 0:
        return n + juft_sonlar_yigindisi(n - 1)
    return juft_sonlar_yigindisi(n - 1)

print(juft_sonlar_yigindisi(10))
```

```python
def royxat_max(royxat):
    if len(royxat) == 1:
        return royxat[0]
    max_qolgan = royxat_max(royxat[1:])
    return royxat[0] if royxat[0] > max_qolgan else max_qolgan

print(royxat_max([4, 9, 2, 7, 5]))
```

```python
def ikkilik_son(n):
    if n == 0:
        return "0"
    if n == 1:
        return "1"
    return ikkilik_son(n // 2) + str(n % 2)
print(ikkilik_son(10))
print(ikkilik_son(7))
```

```python
def royxatda_bormi(royxat, element):
    for son in royxat:
        if son == element:
            return True
    return False

print(royxatda_bormi([3, 7, 1, 9], 7))
print(royxatda_bormi([3, 7, 1, 9], 5)) 
```

```python
def sonlar_kopaytmasi(royxat):
    if len(royxat) == 1:
        return royxat[0]
    return royxat[0] * sonlar_kopaytmasi(royxat[1:])

print(sonlar_kopaytmasi([2, 3, 4]))  
print(sonlar_kopaytmasi([1, 5, 2]))
```
