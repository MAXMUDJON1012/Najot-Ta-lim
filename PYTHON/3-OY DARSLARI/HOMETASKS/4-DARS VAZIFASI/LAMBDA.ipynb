```python
uch_baravar = lambda x: x * 3

print(uch_baravar(4))   
print(uch_baravar(5))   
print(uch_baravar(10)) 
print(uch_baravar(2))   
print(uch_baravar(7))
```

```python
katta = lambda x, y: x if x > y else y

print(katta(8.3,3)) 
print(katta(2.2,9)) 
```

```python
ishora = lambda x: "Manfiy" if x < 0 else ("Nol" if x == 0 else "Musbat")

print(ishora(-5))
print(ishora(0))    
print(ishora(7))
```

```python
yuza = lambda uzunlik, kenglik: uzunlik * kenglik

print(yuza(4,4))  
print(yuza(8,10))
```

```python
def celsius_fahrenheit(celsius):
    f_calc = lambda c: c * 9/5 + 32
    return f_calc(celsius)

print(celsius_fahrenheit(0),celsius_fahrenheit(100))
```
