```python
def narx_chegirma(narx, foiz):
    return narx * (1 - foiz / 100)

print(narx_chegirma(100000, 20))
```

```python
def perimetr(a, b, c):
    return a + b + c

print(perimetr(3, 4, 5))
```

```python
def ortacha_top(royxat):
    return sum(royxat) / len(royxat)
print(ortacha_top([10, 20, 30, 40]))
```

```python
def daraja_chiqar(asos,kursatkich=3):
    return asos*kursatkich
print(daraja_chiqar("salom \n",2))
```

```python
def ekatta_ekichik(a, b, c, d, e):
    royxat = sorted([a, b, c, d, e])
    return royxat[0], royxat[-1]

a, b, c, d, e = map(int, input("Sonlar: ").split())
print(ekatta_ekichik(a, b, c, d, e))
```
