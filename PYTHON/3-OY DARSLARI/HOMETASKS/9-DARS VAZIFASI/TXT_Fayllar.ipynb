```python
with open("ismlar.txt", "w") as f:
    for i in range(1, 6):
        ism = input(f"{i}-ismni kiriting: ")
        f.write(ism + "\n")

with open("ismlar.txt", "r") as f:
    print("\nFayldagi ismlar:")
    print(f.read())
```

```python
try:
    with open("matn.txt", "r") as f:
        qatorlar = f.readlines()
        print("Qatorlar soni:", len(qatorlar))
except FileNotFoundError:
    print("matn.txt fayli topilmadi!")
```

```python
vaqt = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
with open("jurnal.txt", "a") as f:
    f.write(f"Dastur ishga tushdi: [{vaqt}]\n")

with open("jurnal.txt", "r") as f:
    print(f.read())
```

```python
with open("sonlar.txt", "w") as f:
    for son in range(1, 21):
        f.write(str(son) + "\n")

with open("sonlar.txt", "r") as f:
    print("Fayldagi juft sonlar:")
    for qator in f:
        son = int(qator.strip())
        if son % 2 == 0:
            print(son)
```

```python
kalit_soz = input("Qidirilayotgan so'zni kiriting: ")
try:
    with open("matn.txt", "r") as f:
        matn = f.read()
        print(f"'{kalit_soz}' so'zi {matn.count(kalit_soz)} marta uchradi.")
except FileNotFoundError:
    print("matn.txt fayli topilmadi!")
```

```python
with open("talabalar.txt", "w") as f:
    f.write("Ali,85\nVali,72\nZara,91\n")

with open("talabalar.txt", "r") as f:
    print("Bahosi 80 dan yuqori talabalar:")
    for qator in f:
        if qator.strip():
            ism, baho = qator.strip().split(",")
            if int(baho) > 80:
                print(f"{ism} - {baho}")
```

```python
with open("a.txt", "w") as f: f.write("Bu A fayl matni.\n")
with open("b.txt", "w") as f: f.write("Bu B fayl matni.\n")

with open("a.txt", "r") as f1: matn_a = f1.read()
with open("b.txt", "r") as f2: matn_b = f2.read()

with open("natija.txt", "w") as f_natija:
    f_natija.write(matn_a + matn_b)
```

```python
try:
    with open("matn.txt", "r") as f:
        sozlar = f.read().split()
    
    statistika = {}
    for soz in sozlar:
        soz = soz.lower().strip(".,!?")
        statistika[soz] = statistika.get(soz, 0) + 1

    with open("statistika.txt", "w") as f_stat:
        for soz, soni in statistika.items():
            f_stat.write(f"{soz}: {soni}\n")
except FileNotFoundError:
    print("matn.txt fayli topilmadi!")
    
```
