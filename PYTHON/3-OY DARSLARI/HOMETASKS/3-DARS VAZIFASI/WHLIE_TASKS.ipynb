```python
H = float(input("Quduq chuqurligi (H): "))
U = float(input("Kunduzgi ko'tarilish (U): "))
D = float(input("Kechasi sirg'alish (D): "))
joriy_balandlik = 0
kun = 0
while joriy_balandlik < H:
    kun += 1
    joriy_balandlik += U
    if joriy_balandlik >= H:
        break
    joriy_balandlik -= D
print(f"Baqa quduqdan {kun} kunda chiqib ketadi.")
```

```python
S = float(input("Hamyondagi boshlang'ich pul (S): "))
X = float(input("Har safar olinadigan pul (X): "))

marta = 0
if X >= S:
    print("Hamyon 1-marta olishdayoq butunlay bo'shab qoladi (yoki pul yetmaydi).")
elif X <= S / 2:
    print("Pul cheksiz ko'payib ketadi! Hamyon hech qachon bo'shamaydi.")
else:
    while S > 0:
        if S < X:
            print(f"Hamyonda yetarli pul qolmadi. Pul jami {marta} marta olindi.")
            break
        S -= X      
        S *= 2    
        marta += 1  
        if S <= 0:
            print(f"Hamyon {marta} marta pul olingandan keyin butkul bo'shab qoldi.")
            break
```

```python
N = int(input("Shahar aholisi soni (N): "))
K = int(input("1 ta zombi bir kunda nechta odamni tishlaydi (K): "))

zombilar = 1
kun = 0
print(f"Boshlang'ich holat: 1 ta zombi.")
while zombilar < N:
    kun += 1
    yangi_zombilar = zombilar * K
    zombilar += yangi_zombilar
    if zombilar > N:
        zombilar = N
    print(f"{kun}-kuni zombilar soni: {zombilar} ta")
print(f"\nJami {kun} kunda butun shahar aholisi zombiga aylandi.")
```
