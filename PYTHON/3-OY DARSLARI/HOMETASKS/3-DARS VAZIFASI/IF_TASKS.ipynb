```python
A,B,C=map(float,input("Enter the number: ").split())
if A < B < C:
    A = A * 2
    B = B * 2
    C = C * 2
else:
    A = -A
    B = -B
    C = -C
print(f"Natijalar -> A: {A}, B: {B}, C: {C}")
```
