```python
# SET (TO'PLAMLAR)
```

```python
royxat = [1, 2, 2, 3, 4, 4, 5]
natija = set(royxat)
print(natija)
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
natija = a.intersection(b)
print(natija)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}
natija = a.union(b)
print( natija)
```

```python
sozlar4 = ["olma", "banan", "olma", "uzum", "banan"]
takroriy_soni = len(sozlar4) - len(set(sozlar4))
print("Takroriy so'zlar soni:", takroriy_soni)
```

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
natija = a.difference(b)
print(natija)
```

```python
# DICT (LUG'ATLAR)
```

```python
talaba = {"ism": "Ali", "yosh": 18, "baho": 95}
for kalit, qiymat in talaba.items():
    print(f"Kalit: {kalit}, Qiymat: {qiymat}")
```

```python
isimlar=["ali","vali"]
natija={}
for ism in isimlar:
    # natija[ism]=natija.get(ism,0)+1
    if ism in natija.keys():
        natija[ism]+=1
    else:
        natija[ism]=1
print(natija)
```

```python
oylik = {"Ali": 5000000, "Vali": 4500000, "Zara": 6000000}
jami_oylik = sum(oylik.values())
print(jami_oylik)
```

```python
kiritilgan_ismlar = ["Ali", "Vali", "Ali", "Zara", "Ali"]
ismlar_dict = {}
for ism in kiritilgan_ismlar:
    ismlar_dict[ism] = ismlar_dict.get(ism, 0) + 1
print( ismlar_dict)
```

```python
# SET VA DICT ARALASH
```

```python
royxat = [3, 1, 4, 1, 5, 3, 3]
natija = {}
for element in royxat:
    natija[element] = natija.get(element, 0) + 1
print(natija)
```

```python
a= {"ism": "Ali", "yosh": 18}
b = {"shahar": "Toshkent", "baho": 90}
a.update(b)
print(a)
```

```python
baholar = {"Ali": 85, "Vali": 92, "Zara": 78, "Kamol": 90}
eng_katta_kalit = max(baholar, key=baholar.get)
print( eng_katta_kalit)
```

```python
matn = "salom dunyo"
harflar_dict = {}
for harf in matn:
    if harf != " ":
        harflar_dict[harf] = harflar_dict.get(harf, 0) + 1
print(harflar_dict)
```

```python
talabalar = {"Ali": 85, "Vali": 92, "Zara": 95, "Kamol": 78, "Bek": 91}
yaxshi_talabalar = [ism for ism, baho in talabalar.items() if baho > 90]
print(yaxshi_talabalar)
```
