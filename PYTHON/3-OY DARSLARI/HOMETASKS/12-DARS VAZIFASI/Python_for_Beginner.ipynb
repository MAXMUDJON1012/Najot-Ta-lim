```python
from datetime import datetime

tugilgan_sana_str = input("Tug'ilgan sanangizni kiriting: ")
tugilgan_sana = datetime.strptime(tugilgan_sana_str, "%Y-%m-%d").date()
bugun = datetime.now().date()

yashagan_kunlar = (bugun - tugilgan_sana).days

keyingi_tk = datetime(bugun.year, tugilgan_sana.month, tugilgan_sana.day).date()

if keyingi_tk < bugun:
    keyingi_tk = datetime(bugun.year + 1, tugilgan_sana.month, tugilgan_sana.day).date()

qolgan_kunlar = (keyingi_tk - bugun).days

print(f"Siz shu kungacha {yashagan_kunlar} kun yashadingiz.")
print(f"Keyingi tug'ilgan kuningizga {qolgan_kunlar} kun qoldi.")
```
