```python
my_list = [10, 20, 30, 40, 50]
yigindi = 0

for son in my_list:
    yigindi += son

print("Yig'indi:", yigindi)  # 1-usul  
print(sum(my_list))          # 2-usul  
```

```python
my_list = [12, 45, 2, 89, 34]
eng_katta = my_list[0]

for son in my_list:
    if son > eng_katta:
        eng_katta = son

print("Eng katta element:", eng_katta)  # 1-usul
print(max(my_list)) # 1-usul
```

```python
my_list = [12, 45, 2, 89, 34]
eng_kichik = my_list[0]

for son in my_list:
    if son < eng_kichik:
        eng_kichik = son

print("Eng kichik element:", eng_kichik)
print(min(my_list))
```

```python
my_list = [4, 7, 12, 19, 24, 33, 40]
juft_count = 0

for son in my_list:
    if son % 2 == 0:
        juft_count += 1

print("Juft sonlar soni:", juft_count)
sana = len(my_list)
print(sana)
```
