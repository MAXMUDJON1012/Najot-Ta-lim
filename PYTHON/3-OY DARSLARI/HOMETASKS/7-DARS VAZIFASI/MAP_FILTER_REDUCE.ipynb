```python
from functools import reduce
```

```python
sonlar1 = [2, 3, 4]
kublar = list(map(lambda x: x**3, sonlar1))
print(kublar)
```

```python
ismlar = ["Ali", "Vali", "Aziz", "Sardor"]
a_ismlar = list(filter(lambda ism: ism.startswith("A"), ismlar)) #1- usul




a_ismlar1 = list(filter(lambda ism: ism[0]=="A", ismlar))   #2-usul
print(a_ismlar,a_ismlar1)
```

```python
kasr_sonlar = [1.555, 2.349, 3.781]
yaxlitlangan = list(map(lambda x: round(x, 2), kasr_sonlar))
print(yaxlitlangan)
```

```python
sonlar = [10, 15, 20, 30, 45]
bo_linuvchilar = list(filter(lambda x: x % 3 == 0 and x % 5 == 0, sonlar))
print(bo_linuvchilar)
```

```python
sonlar = [7, 2, 9, 1, 5]
eng_kichik = reduce(lambda a, b: a if a < b else b, sonlar)
print(eng_kichik)
```

```python
sozlar = ["salom", "dunyo"]
teskari_sozlar = list(map(lambda s: s[::-1], sozlar))
print(teskari_sozlar)
```

```python
sonlar = [-4, 2, -1, 6, 3, 8]
musbat_juft = list(filter(lambda x: x > 0 and x % 2 == 0, sonlar))
print(musbat_juft)
```

```python
sonlar = [10, 20, 30, 40]
ortacha = reduce(lambda a, b: a + b, sonlar) / len(sonlar)
print(ortacha)
```

```python
sonlar = [11, 22, 33, 44]
qoldiqlar = list(map(lambda x: x % 5, sonlar))
print(qoldiqlar)
```

```python
sonlar = [1, 2, 3, 4]
yigindi = reduce(lambda a, b: a + b, sonlar)
kvadrat = yigindi ** 2
print(kvadrat)
```

```python
sozlar = ["ali", "jasur", "iqbol", "bek"]
katta_va_uzun = list(map(lambda s: s.upper(), filter(lambda s: len(s) > 4, sozlar)))
print(katta_va_uzun)
```

```python
sonlar = [1, 2, 3, 4, 5, 6]
juft_yigindi = reduce(lambda a, b: a + b, filter(lambda x: x % 2 == 0, sonlar))
print(juft_yigindi)
```

```python
talabalar = [("Ali", 90), ("Vali", 78), ("Zara", 95)]
a_talabalar = list(map(lambda t: t[0].upper(), filter(lambda t: t[1] > 85, talabalar)))
print(a_talabalar)
```
