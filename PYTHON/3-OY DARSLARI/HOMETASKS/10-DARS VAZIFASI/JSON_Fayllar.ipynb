```python
import json

mahsulotlar = []
while True:
    nomi = input()
    if nomi == "":
        break
    narxi = int(input())
    mahsulotlar.append({"nomi": nomi, "narx": narxi})

with open("mahsulotlar.json", "w") as f:
    json.dump(mahsulotlar, f, indent=4)
```

```python
import json

with open("mahsulotlar.json", "r") as f:
    mahsulotlar = json.load(f)

eng_qimmat = mahsulotlar[0]
eng_arzon = mahsulotlar[0]

for m in mahsulotlar:
    if m["narx"] > eng_qimmat["narx"]:
        eng_qimmat = m
    if m["narx"] < eng_arzon["narx"]:
        eng_arzon = m

print(f"Eng qimmat: {eng_qimmat['nomi']} — {eng_qimmat['narx']} so'm")
print(f"Eng arzon:  {eng_arzon['nomi']} — {eng_arzon['narx']} so'm")
```

```python
import json

with open("kontaktlar.json", "r") as f:
    kontaktlar = json.load(f)

qidirilayotgan_shahar = input()
topildi = False

for k in kontaktlar:
    if k["shahar"].lower() == qidirilayotgan_shahar.lower():
        print(f"{k['ism']:<10} {k['telefon']}")
        topildi = True

if not topildi:
    print("Bu shaharda kontakt yo'q")
```

```python
import json

with open("a.json", "r") as f: 
    ro_yxat_a = json.load(f)
with open("b.json", "r") as f: 
    ro_yxat_b = json.load(f)

birlashtirilgan = ro_yxat_a.copy()
kiritilgan_ismlar = []

for t in birlashtirilgan:
    kiritilgan_ismlar.append(t["ism"])

for t in ro_yxat_b:
    if t["ism"] not in kiritilgan_ismlar:
        birlashtirilgan.append(t)
        kiritilgan_ismlar.append(t["ism"])

for t in birlashtirilgan:
    print(f"{t['ism']:<5} — {t['baho']}")
```
