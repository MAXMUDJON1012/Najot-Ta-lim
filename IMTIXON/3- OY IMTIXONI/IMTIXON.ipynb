```python
name = input("Ismingizni kiriting: ").strip().title()
while True:

    age= input("Yoshingizni kiriting: ")
    if age.isdigit():
        age = int(age)
        if 0 < age < 120: 
            break
        else:
            print("Iltimos, to'g'ri yosh kiriting (1-120 oralig'ida)!")
    else:
        print("Yosh faqat raqamda bo'lishi kerak!")

if age >= 18:
    print(f"Salom, {name}! Siz voyaga yetgansiz.😈")
else:
    print(f"Salom, {name}! Siz {18 - age} yildan so'ng kotta bo'lasiz.🍼")
```

```python
def sonlar_yigindisi(royxat):
    if not royxat:
        return 0
    return royxat[0] + sonlar_yigindisi(royxat[1:])

print(sonlar_yigindisi([3, 7, 2, 9, 1]))
print(sonlar_yigindisi([10, 20, 30]))
```

```python
sonlar = [-15, 8, -3, 22, -11, 5, 17, -9]
yangi=[]
katta_10=[]
for  n in sonlar:
    if n<0:
        n*= -1
        yangi.append(n)
for i in range(len(sonlar)):
    
    if sonlar [i] in yangi or -sonlar[i] in yangi:
        sonlar[i] = sonlar[i] * -1 
for n in sonlar:
    if n > 10:
        katta_10.append(n)
print(katta_10)
```
